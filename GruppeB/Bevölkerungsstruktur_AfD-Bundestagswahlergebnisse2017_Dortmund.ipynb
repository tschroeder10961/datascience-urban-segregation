#Hypothese: Je höher der Anteil an Ausländern in einem Bezirk ist, desto mehr Menschen wählen AfD
##statistische Hypothese: Es besteht ein positiver Korrelationskoeffizient (r) zwischen den Variablen Anzahle der Ausländer in einem Bezirk und Anzahl der Zweitstimmen für die AfD

```python
#Gesamtanzahl Einwohner 2013
DortmundKomplett['EinwInsg2013'] = DortmundKomplett['Deutsch2013'] + DortmundKomplett['Ausländer2013']
print(DortmundKomplett)
#relative Anzahl Ausländer 2013
DortmundKomplett['RelAusländer2013'] = DortmundKomplett['Ausländer2013'] / DortmundKomplett['EinwInsg2013']
print(DortmundKomplett)
#relative Anzahl Zweitstimmen AfD 2013
DortmundKomplett['RelZweitstimmeAfD'] = DortmundKomplett['AfDZweitstimmen2013'] / DortmundKomplett['GültigeZweitstimmen2013']
(DortmundKomplett).head()
```

```python
#Korrelation RelAusländer2013 und RelZweitstimme2013
KorrelationWahlen2013 = DortmundKomplett['RelAusländer2013'].corr(DortmundKomplett['RelZweitstimmeAfD' ])
print(KorrelationWahlen2013)
```
