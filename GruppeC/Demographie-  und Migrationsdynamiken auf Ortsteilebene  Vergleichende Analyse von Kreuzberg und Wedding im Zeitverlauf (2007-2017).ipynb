# Demographie-  und Migrationsdynamiken auf Orteilsebene 

### Vergleichende Analyse von Kreuzberg und Wedding im Zeitverlauf (2007-2017)

## 1. Einleitung


Während der Teilung Berlins unterlagen die Ortsteile Kreuzberg und Wedding einer ähnlichen sozio-demographischen und städtebaulichen  Entwicklung. Durch verschiedene politische, wirtschaftliche und städtebauliche Interventionen sowie kulturelle und soziale Prozesse, haben sich die Ortsteile nach der Wende jedoch unterschiedlich entwickelt: Während (Friedrichshain-)Kreuzberg sich schon seit einigen Jahren als beliebtes Wohnumfeld für  Kreative, Zugezogene und Studierende etabliert hat, wurde Wedding lange Zeit als "abgehängter" Ortsteil wahrgenommen, welcher durch hohe Arbeitslosigkeit und Kriminalität im Bewusstsein der Berlinerinnen und Berliner stand (Holm, 2016: S.210; von Einem, 2016: S.54). 

Ziel dieser Arbeit ist es die demographische und ethnische Entwicklung zwischen 2007 und 2017 der beiden Ortstteile aufzuzeigen.
Zunächst wird durch deskriptive Methoden überprüft, ob die demographische und ethnische Zusammensetzung der beiden Bezirke sich tatsächlich stark unterscheidet. Darauf aufbauend wird der Segregationsindex auf Ebene der Lebensweltlich orientierten Räume (LOR)  hinsichtlich ethnischer und demografischer  Segregation bestimmt und komparativ zwischen den beiden Ortstteilen betrachtet.
Im letzten Schritt werden etwaige Zusammenhänge zwischen den Veränderungen in der Bevölkerungsentwicklung und den berechneten Segregationindezes mit Hilfe von Korrelationsrechnungen über die Jahre 2007-2017 ergründet.



## 2. Forschungsfragen:

Die Forschungsfragen strukturieren sich anhand der Vorgehensweise dieser Arbeit. Die erste Forschungsfrage dient der deskriptiven Untersuchung der vorliegenden Daten hinsichtlich der Sozialstruktur. In der zweiten Forschungsfrage wird sowohl die demographische als auch die ethnische Segregation untersucht. Die soziale Segregation, welche die räumliche Polarisierung anhand sozio-ökonomischer Kriterien wie Einkommen oder sozialem Status beschreibt, wird in dieser Arbeit aufgrund von unzureichendem Datenmaterial auf LOR-Ebene ausgeklammert.
Um eine Aussage bezüglich der migrantischen Entwicklung der Ortsteile treffen zu können, werden sowohl Personen mit Migrationshintergrund als auch Ausländer berücksichtigt, um eine umfassender Aussage bzgl. der ethnischen Segregation treffen zu können. Personen mit Migrationshintergrund unterscheiden sich durch den Besitz der deutschen Staatsbürgerschaft und den damit verbundenen Rechten wie beispielsweise des Wahlrechts von der Gruppe der Ausländer, deren Rechte im Aufenthaltsrecht geregelt sind.

Die unter 15 Jährigen wurden ebenfalls berücksichtigt und dienen der Erfassung der Segregation bzw. der Anzahl der Familien in den jeweiligen Ortsteilen (siehe Helbig und Jähnen, 2018).

## Beantwortung Forschungsfrage 2

Wie hoch ist die ethnische und demographische Segregation auf Stadtteilebene für Kreuzberg & Wedding im Zeitverlauf?

Um eine Aussage hinsichtlich der  ethnischen Segregation treffen zu können, werden die Indizes für die Bevölkerungsgruppe der Personen mit Migrationshintergrund und Ausländer berücksichtigt. S

Die demographische Segregation wird hier verstanden als die räumliche Polarisierung der Personen über 65 und unter 15.

## 1. Vorgehen
Für den Segregationsindex wird zunächst der Anteil der zu betrachtenden Bevölkerungsgruppe zur Gesamtbevölkerung bestimmt und in ein Verhältnis zu dem Restanteil an der Gesamtbevölkerung im Stadtbezirk gesetzt. Der daraus resultierende Absolutbetrag wird über die LORs aufsummiert und mit dem Faktor 0,5 multipliziert, woraus sich der Segregationsindex für den Stadtteil im jeweiligen Jahr ergibt. Der Segregationsindex liegt zwischen 0 und 1 und gibt an wie viele Menschen in dem Ortsteil umziehen müssten, um eine absolute Gleichverteilung zu erreichen.

## Zusammenfassung


Im Hinblick auf die Entwicklungen der beiden Ortsteile Wedding und Kreuzberg lassen sich bezüglich der demographischen sowie ethnischen Zusammensetzung in der Bevölkerungsentwicklung einige Unterschiede erkennen. 

#### Demographische Entwicklungen

Es konnte gezeigt werden, dass im Hinblick auf die Gruppe der unter 15-Jährigen in den beiden Stadtteilen über die letzten zehn Jahre keine einschneidenden Veränderungen aufgetreten sind und sich der Anteil in den beiden Stadtteilen ungefähr auf einem ähnlichen Niveau befindet. In der Gruppe der 65-Jährigen zeigt sich ein Trend der Annäherung zwischen den beiden Stadtteilen. Während der Anteil in Wedding von 14,2 % auf 12,3 % gesunken ist, hat sich der Anteil der über 65 Jährigen in Kreuzberg um einen Prozentpunkt auf 10,3% erhöht. 
Im Hinblick auf die Segregation in den beiden Stadtteilen zeigt sich in Kreuzberg, dass die demographische Segregation deutlich abgenommen hat. Während in 2007 noch 11,6% der Menschen über 65 Jahre hätten umziehen müssen, um eine gleiche demographische Verteilung innerhalb des Stadtteils zu erlangen, liegt dieser Wert im Jahr 2017 nur noch bei 6,4%. Die Segregationsindizes der letzten zehn Jahre für die Gruppe der unter 15-Jährigen validieren diese Entwicklung mit einem Absinken des Segregationsindex von 9,5% auf 6,1%.  In Wedding ist in der Gruppe der unter 15-Jährigen ebenfalls eine Verkleinerung des Segregationsindex zu beobachten (von 10,9% auf 8,8 %), während der Index für die über 65-Jährigen relative konstant bei ca. 15 % geblieben ist. 
Wie die berechneten Segregationsindizes über den Untersuchungszeitraum hinweg mit den Anteilen in Zusammenhang stehen, konnte mithilfe von Korrelationsberechnungen ermittelt werden.  Besonders in Kreuzberg zeigte sich der Zusammenhang zwischen den Anteilen der beiden Altersgruppen und ihren jeweiligen Segregationswerten. Mit einem steigenden Anteil von über 65-Jährigen, verkleinert sich Segregationsindex (r = -0.853), während bei einem steigenden Anteil von unter 15-Jährigen ein größerer Segregationsindex vorliegt (r = 0.843).  In Wedding ist dieser Trend ebenfalls zu beobachten, auch wenn die Korrelationsmaße mit r = -0,310 in der Gruppe der über 65-Jährigen und r = 0.843 bei den unter 15-Jährigen etwas niedriger ausfallen. Außerdem kann in Wedding nicht von linearen Zusammenhängen gesprochen werden, in Kreuzberg hingegen schon. In beiden Stadtteilen verkleinert sich der demographische Segregationsindex bei Anteilszuwachs der über 65-Jährigen und vergrößert sich bei Anteilszuwachs der über 15-Jährigen.


#### Ethnische Entwicklungen

Die Bevölkerungsgruppen der Ausländer und der Personen mit Migrationshintergrund weisen ebenfalls Unterschiede in den beiden Stadtteilen auf. 
Mit einem relativ konstanten Anteil von ca. 30 % in den letzten zehn Jahren liegt der Anteil der Ausländer deutlich unter dem Anteil in Wedding. Dort ist dieser insgesamt von 33,9 % auf 36,5 % angestiegen. Ein ähnlicher Trend zeigt sich für Wedding auch bei dem Anteil der Menschen mit Migrationshintergrund, der von 53 % auf 58 % kontinuierlich angestiegen ist. In Kreuzberg hingegen ist dieser relativ konstant bei circa 50 % geblieben mit einem jüngsten Rückgang von 52 % auf 48,3 %. 
Der Segregationsindex über den Untersuchungszeitraum weist in der Gruppe der Ausländer in Kreuzberg ebenso wie der Gesamtanteil nur eine geringe Veränderung von 10 % auf 9,3 %.  Auch bei der Gruppe der Menschen mit Migrationshintergrund bleibt der Segregationsindex in Kreuzberg bei einem relativ konstanten Wert, wohingegen diese Gruppe in Wedding eine Verkleinerung von 18,1 % auf 13,6 % vorzuweisen hat. 
Auch bei dem Segregationsindex der Ausländer zeigt sich in Wedding eine Verkleinerung von 15,1 % auf 12 % zum Ende des Untersuchungszeitraums. 
Ein Zusammenhang zwischen dem Ausländeranteil und dem Segregationsindex konnte in beiden Stadtteilen eindeutig nachgewiesen werden. Mit einem Anteilszuwachs verkleinert sich der Index jeweils (r = -0.803 in Kreuzberg und r = -0.579 in Wedding). Im Hinblick auf den Zusammenhang zwischen dem Anteil und dem Segregationsindex der Gruppe der Menschen mit Migrationshintergrund zeigt sich lediglich bei Wedding ein starker Zusammenhang (r = -0.845). In Kreuzberg kann von keinem Zusammenhang gesprochen werden. 


Insgesamt zeigt sich, dass im Bezug auf die demographischen Unterschiede zwischen den beiden Stadtteilen, dass in Kreuzberg der Trend zu einer gleichmäßigen Verteilung der Altersgruppen erfolgt ist. Wohingegen in Wedding die Segregation besonders bei den über 65-Jährigen eine höhere Segregation aufweiset.
In Bezug auf die ethnische Segregation bleiben sowohl die Anteile als auch die Indizes der beiden Gruppen auf einem annähernd gleichen Niveau. Wedding hat nicht nur einen relativ deutlichen Anteilszuwachs zu verzeichnen,  sondern zeigt auch den Trend zu einer gleichmäßigeren Verteilung der Bevölkerungsgruppen über die einzelnen LOR. Dabei liegt der Segregationsindex der Ausländer in Wedding trotz eines Anteilszuwachs knapp über dem in Kreuzberg, während sich die Personen mit Migrationshintergrund in Wedding mit einem 8 %-höheren Anteil dieser Gruppe deutlich besser im Stadtteil verteile,n als es in Kreuzberg der Fall ist.
Des Weiteren lässt sich in beiden Städten beobachten, dass ein Bevölkerungsrückgang um die Jahre 2009 bis 2011 stattgefunden hat. Dieser spiegelt sich auch in der Bevölkerungsentwicklung in einer der Gruppen wider. Über den Untersuchungszeitraum hinweg zeigen beide Stadtteile einen Tiefststand im Jahr 2010 bei dem Anteil der Ausländer.
